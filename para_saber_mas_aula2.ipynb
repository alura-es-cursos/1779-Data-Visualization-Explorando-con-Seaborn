```
import pandas as pd
import seaborn as sns
```

```
df = pd.read_csv('credit_card.csv')
```

```
df.head()
```

```
primer_grafico = sns.scatterplot(x='AGE', y='BILL_AMT', data=df);
```

```
primer_grafico
```

```
primer_grafico.get_figure()
```

**figure.suptitle**

```
primer_grafico.figure.suptitle('Edad x Valor de la factura')
primer_grafico.get_figure()
```

**set_title**

```
primer_grafico.set_title('Anlisis de la edad y el valor de la factura')
primer_grafico.get_figure()
```

**Alterar labels**

```
primer_grafico.set(xlabel='Edad', ylabel='Valor de la factura')
primer_grafico.get_figure()
```

**Descargar imagen**

```
imagen = primer_grafico.get_figure()
```

```
imagen.savefig('imagen.png')
```
